### Regular Linear Regression

```python
from sklearn.model_selection import KFold
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_absolute_percentage_error, mean_squared_error, r2_score
```

```python
class LinearRegression:
    def __init__(self):
        self.w = None
        
    def fit(self, X, y):
        lenw = X.shape[1]
        numK = X.shape[0]
        w = cp.Variable(lenw)
        residuals = y - X @ w
        objective = cp.Minimize(cp.sum(cp.square(cp.abs(residuals))))
        prob = cp.Problem(objective)
        prob.solve()
        self.w = w.value

    def predict(self, X):
        return X @ self.w
```

```python
N_SPLITS = 5
RESULT_METRICS = {}
kf = KFold(n_splits=N_SPLITS, shuffle=True, random_state=125)

y = df["MEDV"].to_numpy()
X = df.drop("MEDV", axis=1).to_numpy()
```

```python
avg_mape = 0
avg_mse = 0
avg_r2 = 0
for i, s in enumerate(kf.split(df)):
    X_train = X[s[0]]
    y_train = y[s[0]].reshape(-1, 1)
    X_test = X[s[1]]
    y_test = y[s[1]].reshape(-1, 1)

    X_scaler = StandardScaler()
    X_train = X_scaler.fit_transform(X_train)
    X_test = X_scaler.transform(X_test)
    y_scaler = StandardScaler()
    y_train = y_scaler.fit_transform(y_train).ravel()
    y_test = y_scaler.transform(y_test).ravel()

    lr = LinearRegression()
    lr.fit(X_train, y_train)
    y_pred = lr.predict(X_test)
    print(f'Split {i}')
    print(f'\tMAPE={mean_absolute_percentage_error(y_test, y_pred)}')
    print(f'\tMSE={mean_squared_error(y_test, y_pred)}')
    print(f'\tR2={r2_score(y_test, y_pred)}')
    print()
    avg_mape += mean_absolute_percentage_error(y_test, y_pred)
    avg_mse += mean_squared_error(y_test, y_pred)
    avg_r2 += r2_score(y_test, y_pred)
avg_mape /= N_SPLITS
avg_mse /= N_SPLITS
avg_r2 /= N_SPLITS

print(f'Average MAPE={avg_mape}')
print(f'Average MSE={avg_mse}')
print(f'Average R2={avg_r2}')

RESULT_METRICS["Linear Regression"] = {}
RESULT_METRICS["Linear Regression"]["mape"] = avg_mape
RESULT_METRICS["Linear Regression"]["mse"] = avg_mse
RESULT_METRICS["Linear Regression"]["r2"] = avg_r2
```

### Try robust implementations: Ridge, HuberRegressor, Lasso

```python
class LassoLR:
    def __init__(self, delta):
        self.w = None
        self.delta = delta
        
    def fit(self, X, y):
        lenw = X.shape[1]
        numK = X.shape[0]
        w = cp.Variable(lenw)
        residuals = y - X @ w
        objective = cp.Minimize(cp.sum(cp.square(cp.abs(residuals))) + self.delta * cp.norm(w, 1))
        prob = cp.Problem(objective)
        prob.solve()
        self.w = w.value

    def predict(self, X):
        return X @ self.w
    
class RidgeLR:
    def __init__(self, delta):
        self.w = None
        self.delta = delta
        
    def fit(self, X, y):
        lenw = X.shape[1]
        numK = X.shape[0]
        w = cp.Variable(lenw)
        residuals = y - X @ w
        objective = cp.Minimize(cp.sum(cp.square(cp.abs(residuals))) + self.delta * cp.norm(w, 2))
        prob = cp.Problem(objective)
        prob.solve()
        self.w = w.value

    def predict(self, X):
        return X @ self.w
    
class HuberLR:
    def __init__(self, delta):
        self.w = None
        self.delta = delta
        
    def fit(self, X, y):
        lenw = X.shape[1]
        numK = X.shape[0]
        w = cp.Variable(lenw)
        residuals = y - X @ w
        objective = cp.Minimize(cp.sum(cp.huber(residuals, self.delta)))
        prob = cp.Problem(objective)
        prob.solve()
        self.w = w.value

    def predict(self, X):
        return X @ self.w

import cvxpy as cp
import numpy as np

class HingeLR:
    def __init__(self, delta):
        self.w = None
        self.delta = delta

    def fit(self, X, y):
        n_features = X.shape[1]
        self.w = cp.Variable(n_features)
        
        # Define hinge loss: max(0, |y - Xw| - epsilon)
        epsilon = 1.0  # Tolerance for error (can be tuned)
        residuals = cp.abs(y - X @ self.w)
        hinge_loss = cp.sum(cp.maximum(0, residuals - epsilon))
        
        # Define the objective function with regularization
        objective = cp.Minimize(hinge_loss + self.delta * cp.norm(self.w, 2))
        prob = cp.Problem(objective)
        prob.solve()

    def predict(self, X):
        return X @ self.w.value
```

### Robust Linear Regression

```python
class RobustLR:
    def __init__(self, delta):
        self.w = None
        self.delta = delta
        
    def fit(self, X, y):
        lenw = X.shape[1]
        numK = X.shape[0]
        w = cp.Variable(lenw)
        residuals = y - X @ w
        objective = cp.Minimize(cp.sum(cp.square(cp.abs(residuals) + self.delta * cp.norm(w, 2))))
        prob = cp.Problem(objective)
        prob.solve()
        self.w = w.value

    def predict(self, X):
        return X @ self.w
```

```python
DELTA = 1e-2

avg_mape = 0
avg_mse = 0
avg_r2 = 0

for i, s in enumerate(kf.split(df)):
    X_train = X[s[0]]
    y_train = y[s[0]].reshape(-1, 1)
    X_test = X[s[1]]
    y_test = y[s[1]].reshape(-1, 1)

    X_scaler = StandardScaler()
    X_train = X_scaler.fit_transform(X_train)
    X_test = X_scaler.transform(X_test)
    y_scaler = StandardScaler()
    y_train = y_scaler.fit_transform(y_train)
    y_test = y_scaler.transform(y_test)

    lr = RobustLR(delta=DELTA)
    lr.fit(X_train, y_train.ravel())
    y_pred = lr.predict(X_test)
    print(f'Split {i}')
    print(f'\tMAPE={mean_absolute_percentage_error(y_test, y_pred)}')
    print(f'\tMSE={mean_squared_error(y_test, y_pred)}')
    print(f'\tR2={r2_score(y_test, y_pred)}')
    print()
    avg_mape += mean_absolute_percentage_error(y_test, y_pred)
    avg_mse += mean_squared_error(y_test, y_pred)
    avg_r2 += r2_score(y_test, y_pred)
avg_mape /= N_SPLITS
avg_mse /= N_SPLITS
avg_r2 /= N_SPLITS

print(f'Average MAPE={avg_mape}')
print(f'Average MSE={avg_mse}')
print(f'Average R2={avg_r2}')

RESULT_METRICS["Robust LR"] = {}
RESULT_METRICS["Robust LR"]["mape"] = avg_mape
RESULT_METRICS["Robust LR"]["mse"] = avg_mse
RESULT_METRICS["Robust LR"]["r2"] = avg_r2
```

```python
DELTA = 1e-1

models = {"Linear Regression": LinearRegression,
          "Robust LR": RobustLR,
          "Huber": HuberLR,
          "Lasso": LassoLR,
          "Ridge": RidgeLR}

for model_name, model in models.items():
    np.random.seed(125)
    avg_mape = 0
    avg_mse = 0
    avg_r2 = 0
    for i, s in enumerate(kf.split(df)):
        X_train = X[s[0]]
        y_train = y[s[0]]
        X_test = X[s[1]]
        y_test = y[s[1]]

        X_scaler = StandardScaler()
        X_train = X_scaler.fit_transform(X_train)
        X_test = X_scaler.transform(X_test)
        y_scaler = StandardScaler()
        y_train = y_scaler.fit_transform(y_train.reshape(-1, 1)).ravel()
        y_test = y_scaler.transform(y_test.reshape(-1, 1)).ravel()

        N = np.random.normal(0, 1, X_train.shape)
        row_norms = np.linalg.norm(N, ord=2, axis=1, keepdims=True)
        D = np.diag (np.reciprocal(row_norms).ravel())
        N = np.matmul(D, N)
        X_train = X_train + N * NOISE_DELTA

        N = np.random.normal(0, 1, X_test.shape)
        row_norms = np.linalg.norm(N, ord=2, axis=1, keepdims=True)
        D = np.diag (np.reciprocal(row_norms).ravel())
        N = np.matmul(D, N)
        X_test = X_test + N * NOISE_DELTA
        
        if model_name == "Linear Regression":
            lr = model()
        else:
            lr = model(DELTA)
        lr.fit(X_train, y_train)
        y_pred = lr.predict(X_test)

        avg_mape += mean_absolute_percentage_error(y_test, y_pred)
        avg_mse += mean_squared_error(y_test, y_pred)
        avg_r2 += r2_score(y_test, y_pred)
    avg_mape /= N_SPLITS
    avg_mse /= N_SPLITS
    avg_r2 /= N_SPLITS

    print(model_name)
    print(f'Average MAPE={avg_mape}')
    print(f'Average MSE={avg_mse}')
    print(f'Average R2={avg_r2}')
    print()

    RESULT_METRICS[model_name]["noisy_mape"] = avg_mape
    RESULT_METRICS[model_name]["noisy_mse"] = avg_mse
    RESULT_METRICS[model_name]["noisy_r2"] = avg_r2
```

```python
# Prepare data for visualization
from sklearn.model_selection import KFold

def plot_distribution(model, X, y, kf, x_label="", y_label="", title=""):
    split = next(kf.split(X))
    X_train, y_train = X[split[0]], y[split[0]]
    X_test, y_test = X[split[1]], y[split[1]]
    
    # Масштабирование данных
    scaler_X = StandardScaler()
    scaler_y = StandardScaler()
    X_train_scaled = scaler_X.fit_transform(X_train)
    y_train_scaled = scaler_y.fit_transform(y_train.reshape(-1, 1)).ravel()
    X_test_scaled = scaler_X.transform(X_test)
    y_test_scaled = scaler_y.transform(y_test.reshape(-1, 1)).ravel()
    
    # Обучение модели
    model.fit(X_train_scaled, y_train_scaled)
    
    # Предсказания
    y_pred_scaled = model.predict(X_test_scaled)
    y_pred = scaler_y.inverse_transform(y_pred_scaled.reshape(-1, 1))
    
    # Визуализация
    plt.figure(figsize=(10, 6))
    
    # Исходные точки данных
    plt.scatter(X_train, y_train, label="Train Data", color="blue", alpha=0.6)
    plt.scatter(X_test, y_test, label="Test Data", color="orange", alpha=0.8)
    
    # Аппроксимирующая линия
    X_full = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
    X_full_scaled = scaler_X.transform(X_full)
    y_full_scaled = model.predict(X_full_scaled)
    y_full = scaler_y.inverse_transform(y_full_scaled.reshape(-1, 1))
    
    plt.plot(X_full, y_full, label="Regression Line", color="red", linewidth=2)

    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.legend()
    plt.title(title)
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
df = pd.read_csv('/kaggle/input/boston-housing-dataset/HousingData.csv')
df = df.dropna()

y = df["MEDV"].to_numpy()
X = df[["AGE"]].to_numpy()  # Используем одну фичу для простоты визуализации

# Разделение данных
kf = KFold(n_splits=5, shuffle=True, random_state=125)
delta = 1e-2
models = {"Linear Regression": LinearRegression,
          "Robust LR": RobustLR,
          "Huber": HuberLR,
          "Lasso": LassoLR,
          "Ridge": RidgeLR}

for model_name, model in models.items():
    lr = model() if model_name == "Linear Regression" else model(delta)
    plot_distribution(lr, X, y, kf, x_label="RM (Average number of rooms per dwelling)", \
                      y_label="MEDV (Median value of owner-occupied homes in $1000's)", title=model_name)

```
